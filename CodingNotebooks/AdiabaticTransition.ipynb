## **Introduction**

In this assignment, you will be exploring how adiabatic transitions affect the spin state of particles. As discussed in lecture and lab, an adiabatic transition is a way of preserving the original state of a particle while changing its orientation.

Another way to describe this is to imagine that the positive z-direction points upward toward the sky in the lab frame. Through an adiabatic transition, an experimenter can pass a particle through a series of Stern–Gerlach (SG) devices whose orientations rotate gradually from directly upward to the right, which we can define as the positive x-direction.

If this measurement were performed using only one SG device aligned in the z-direction to select the positive z spin state, followed by a second SG device aligned in the positive x-direction, one would expect a fifty–fifty split between positive and negative x spin outcomes. This is because measuring the particle’s spin in the x-direction destroys any prior information about its spin in the z-direction. As a result, a measurement in the x-direction yields equal probabilities for spin-up and spin-down.

However, if instead of making an abrupt change in orientation, the experimenter aligns a series of SG devices such that each device differs only slightly in angle from the previous one, there will be an increased probability of measuring the particle in the positive x spin state. One can then imagine making these angular changes infinitesimally small. In the adiabatic limit, the probability of measuring a positive x spin remains essentially one hundred percent, even when the final SG device is aligned along the x-direction. 

Notice the adiabatic process preserves the information about the particles spin being positive or negative, not the direction it is measured in. This allows the experimenter to construct a beam of z+ particles and rotate them to the x+ direction and still measure one hundered percent of the original particles. This is all assuming the rotation angle of the intermediate devices are infinitly small.
