## **Introduction**

In this notebook you will explore the behavior of driven damped harmonic oscillators. As observed in class these types of oscillators have a transient and steady state component. The transient part of the solution refers to the motion of the oscillator dictated by its initial conditions, that being position and velocity. The transient portion of the motion will eventually dissapear due to energy loss from damping. This will look different depending on whether the oscillator is over, under, or critically damped. The transient part is best visualized as the part of the motion before steady oscillations begin. The steady state of the motion is when the motion due to initial conditions dissapears and steady oscillations from the driving force begin. These often take the shape of sinusoidal waves.

As seen in class, the differential equations governing these driving forces rely on the driving force to be periodic. More specifically to solve the ODE the driving force needs to be a sine or cosine wave. What we will see by the end of this is how to model the motion of an oscillator experiencing a periodic driving force that is not a sine or cosine wave.

## Definitions
Definitions are integral parts of code allowing the coder to define a specific operation to be used again and again. This is especially important when the coding project starts to become thousands of lines long and span across multiple different files. Definitions have four main parts, the name, variables, functionality, and output. 

#### Name
The name is what you will need to type out later to use the definition: `def Function_name():`. Think of this like all of the other functions you use inside of numpy and matplotlib. For example, when you type `plt.plot()` you are using a function defined by matplotlib; here I used an assumed abbreviation of the matplotlib.pyplot import and called it plt. This same structure can be used if you have a coding project that spans across multiple files. Say you make a definition `def My_Func():` in one file called `"Func.py`" and want to use it in a different file, all you would need to do is open this second file and type `Func.My_Func()`; here the `Func` references the file you want to look into and the `My_Func` is the thing you want to access inside this seperate file. If you are working in the same file you created the function in you can just use it by typing `My_Func()`. All this being said, defining functions is very handy for larger coding projects and makes them run much faster too.

#### Variables
After defining the function it will need variables to take in `def My_Func(Var1, var2, var3):`. These varibles do not necesarrily need to be ints or floats and could be strings. As you may know in the `plt.plot` function you can change the color of the line using a string `plt.plot('b')`. similarly anything you want to be used as an input in your function must be defined here.

#### Functionionality
Now that the function has a name and inputs you need to define what exactly you would like it to do. This is the main body of a function and it comes after the colon in the first line and is indented: 
``` python
def My_Func(var1, var2, var3):
    var1 + var2
    var3
    #add whatever you want your function to do here
```
For the purposes of this notebook this will be where you write the code to model the damped driven oscillators.

#### Output
Once you are finished writing the functionallity of the function you need to tell the computer what of all the things you wrote you want the output of the function to be. This is denoted by `return 'whatever'`. Going back to the example above say I want to print the output of `var1 + var2` I would edit the definition as follows:
```python
def My_Func(var1, var2, var3):
    var1 + var2 = result
    var3
    return result
```
In this code snippet I have made a function that takes in there inputs and returns the sum of the first two. Note the third variable is not mentioned in the return so as of right now it does nothing. Below is an example of the output of this function.

## **Example of using a Definition to make a plotable function**
For this notebook you will be asked to make definitions for the motion of the oscillators so here I will provide a simple example of how to use a definition to make a plotable graph.

````python
import numpy as np
import matplotlib.pyplot as plt

def sin_wave(t, Amplitude, frequency, phase):  #Here I have defined my variables. These are all that's needed to define Sine
    x = Amplitude * np.sin(frequency * t + phase)     #Here I have made a variable that will tell me the x-position of a sine wave
    # Notice I have used all of the variable I defined in my function in the main body of my function
    return x  #This will output the value of my sine wave

'''
If I want a single value out of my function I only put in one value for each variable.
'''
print( sin_wave(1, 1, np.pi/2, 0) )  #This is used to see if my function works. I tested what sin(pi/2) was which should equal 1

'''
If instead I want to see the position of the sine wave at multiple times I will need to enter multiple times
'''
t = np.linspace(0,10, 100) # This makes a list of times from 0-10 with 100 steps between them.
print( sin_wave(t, 1, np.pi/2, 0) )  #This time when I print I get a value of sine for each of the times I input.

'''
If I now want to graph this sine wave I can plot it as follows
'''
plt.plot(t, sin_wave(t, 1, np.pi/2, 0), label='Sine Wave' )
plt.legend()
plt.xlabel("Time (s)")
plt.ylabel('X Position (m)')
plt.title('Example of Sine Function')
plt.show()
````
