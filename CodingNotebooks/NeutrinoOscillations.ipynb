## **Introduction**


Neutrinos are elementary particles with no charge, and very small mass, much smaller than that of an electron. The neutrinos come in three different "flavors," the electron, muon, and tau. For convienence we will only be considering the electron and muon neutrinos as this greatly simplifies the math and keeps the spirit of neutrino oscillations.

The electron neutrino, $\nu_e$, and the muon neutrino, $\nu_{\mu}$, are both comprised of eigenstates of the neutrinos energy. These eigenstates, $E_1$ and $E_2$, corespond to the "mass eigenstates" $m_1$ and $m_2$ which might be denoted at $\lvert \nu_1 \rangle$ and $\lvert \nu_2 \rangle$ respectively. If we were considering the tau neutrino there would be a third energy and mass eigenstate. One can write the $\nu_e$ and $\nu_{\mu}$ states as follows

$$
\lvert \nu_e \rangle = \cos\theta \, \lvert \nu_1 \rangle - \sin\theta \, \lvert \nu_2 \rangle
$$

$$
\lvert \nu_{\mu} \rangle = \sin\theta \, \lvert \nu_1 \rangle - \cos\theta \, \lvert \nu_2 \rangle
$$

These formulas are different than those in McIntyre and instead come from Sakurai. For more information on neutrino oscillations I urge you to read Sakurai pg. 71-73. Notice if the mixing angle $\theta$ was zero, than $\lvert \nu_e \rangle$ and $\lvert \nu_{\mu} \rangle$ would be $\lvert \nu_1 \rangle$ and $\lvert \nu_2 \rangle$ respectively. This mixing angle though had no theoretical value and must be resolved through experiment.

**Why do Neutrinos Oscillate?**

Neutrinos are produced and detected in flavor eigenstates, but they propagate as a superposition of mass eigenstates with different masses. As the neutrino travels, each mass eigenstate acquires a different quantum phase due to its mass. These phase differences lead to interference between the mass eigenstates, causing the neutrinoâ€™s flavor content to change with distance. When the neutrino is detected, the quantum state collapses into a specific flavor with a probability determined by the relative phases of the mass eigenstates.

**Calculating Measurment Probabilities**

Since neutrinos interacte very weakly with matter we can treat their energy as that of a free particle in space. Neutrinos are also highly relativistic particles given, their very small mass and their near light speed travel, meaning their energy will be goverened by special relativity. A good approximation of the energy eigenvalue of a neutrino with fixed momentum $p$ and mass $m$ is

$$
E = [p^2c^2 + m^2c^4]^{\frac{1}{2}} \approx pc(1 + \frac{m^2c^2}{2p^2})
$$

If one starts with a beam of pure $\lvert \nu_e \rangle$ and allows it to evolve, and then at some later time t ask what is the probability of still measuring $\lvert \nu_e \rangle$, one finds

$$
P(\nu_e -> \nu_e) = 1 - \sin^2 2\theta\sin^2(\Delta m^2c^4\frac{L}{4E\hbar c})
$$

where $\Delta m^2 = m^2_1 - m^2_2$. For this assignment please you the following values:<br>
c = 299792458 ($\frac{m}{s}$)<br>
$\hbar$ = $1.054571817$ x $10^{-34}$ ($J \cdot s$)<br>
E = a range from 4 MeV - 8 MeV<br>
L = a range from 0 km - 1300 km<br>
$\Delta m^2$ = $8$ x $10^{-5}$ $eV^2$<br>
$\theta$ = $33\degree$
